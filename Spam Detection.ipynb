```python
sw = stopwords.words('english')
lemma = WordNetLemmatizer()
```

```python
def textcleaner (text):
    assert(type(text)==str)
    text = text.lower()
    cleaned = ' '.join([lemma.lemmatize(word)for word in text.split() if word not in sw])
    return cleaned
```
